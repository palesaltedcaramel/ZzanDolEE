```python
#위도 경도를 추출해주는 사용자 정의 함수 만들기
!pip install geopy
import geopy
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x = geo.latitude 
    y = geo.longitude
    return x, y
```

```python
#위도, 경도 채우기
for district in districts:
    idx =0
    for i in district['세부 주소']:
        try:
            district.loc[idx,'위도'] = geocoding(i)[0]
            district.loc[idx,'경도'] = geocoding(i)[1]
        except:
            pass
        idx += 1
    district
```

```python
#흡연시설형태 표준화

df['흡연시설형태'].unique()

clean_type = df['흡연시설형태'].apply(str).tolist()

patterns1 = ['밀폐형 실외 흡연실','밀폐형', '완전폐쇄형']
patterns2 = ['개방형 실외 흡연실','완전개방형']

for pattern in patterns1:
    clean_type = [re.sub(pattern,'폐쇄형', types) for types in clean_type]
    
for pattern in patterns2:
    clean_type = [re.sub(pattern,'개방형', types) for types in clean_type]

clean_type

df['흡연시설형태'] = clean_type
```

```python
#nan값 확인, 제거 
df.dropna(inplace = True)
```

```python
#정규식으로 주소 clean
r1 = '\([가-힣]*[로][ ][\d]*\)' #'골든씨티호텔(천호대로 377)1층'
r2 = '\([가-힣]*[로]\)*' #~~로
r3 = '\([가-힣]*[동]\)*' #~~동
r4 = '\(주\)|옆|뒤편|10개소|\)|옥상|주차장|내|추자창|후문|일부|비+[\d]|지층'
patterns = [r1,r2,r3,r4]
```

```python
clean_address = df['세부 주소_clean'].apply(str).tolist()

for pattern in patterns:
    clean_address = [re.sub(pattern,'', address) for address in clean_address]
    
clean_address
```

```python
df['세부 주소_clean'] = clean_address
```
