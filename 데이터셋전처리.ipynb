#### 강서구

```python


gangseo = pd.read_csv('./강서구/서울특별시 강서구_실외 흡연 장소_20201008.csv', encoding = 'euc-kr')
gangseo
```

```python
import re

#불필요한 컬럼 삭제
gangseo = gangseo.drop(['시설 구분', '규모', '설치일', '관리여부', '관리'], axis = 1)

#시설형태 표준화
gangseo['시설형태'] = gangseo['시설형태'].replace('완전' ,'', regex = True)
gangseo

#설치위치 컬럼이름을 세부 주소로
gangseo.rename(columns = {'자치구' : '지역(행정구)', '설치 위치': '세부 주소',  '설치 주체': '건물/상호명'}, inplace = True)

#세부주소에 서울특별시 강서구 붙이기
idx = 0
for i in list(gangseo[ '세부 주소']):
    i = '서울특별시 강서구 ' + str(i)
    gangseo.loc[idx, '세부 주소'] = i
    idx += 1
    
    
gangseo
```

```python
#주소 표기 정규표현식

p = re.compile('([\w]+[시])')
p1 = re.compile('[\w]+[구]')
p2 = re.compile ('[\w]+[로]|[\w]+[길]')
p3 = re.compile('[\d]+')
print(re.search(p, str(gangseo.loc[4, '세부 주소'])).group())
print(re.search(p1, str(gangseo.loc[4, '세부 주소'])).group())
print(re.search(p2, str(gangseo.loc[4, '세부 주소'])).group())
print(re.search(p3, str(gangseo.loc[4, '세부 주소'])).group())
```

```python
#주소 다듬기
idx = 0
for i in list(gangseo[ '세부 주소']):
    p1 = re.compile('([\w]+[시])')
    p2 = re.compile('[\w]+[구]')
    p3 = re.compile ('[\w]+[로]')
    p4 = re.compile ('[\w]+[길]')
    p5 = re.compile('[\d]+')
    
    re1 = re.search(p1, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    re2 = re.search(p2, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    try:
        re3 = re.search(p3, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    except:
        re3 = ''
    try:
        re4 = re.search(p4, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    except:
        re4 = ''
    re5 = re.search(p5, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    
    gangseo.loc[idx, '세부 주소_clean'] = re1+re2+re3+re4+re5
    idx += 1
gangseo
```

```python
#주소 다듬기 2
idx = 0
for i in list(gangseo[ '세부 주소']):
    p1 = re.compile('([\w]+[시])')
    p2 = re.compile('[\w]+[구]')
    p3 = re.compile ('([\w]+[로]) ([\w]+[길] [\d]+)')       # ~로~길이 붙어있음;;;;
    
    p4 = re.compile ('[\w]+[로] [\d]+')
    p5 = re.compile ('[\w]+[길] [\d]+')

    re1 = re.search(p1, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    re2 = re.search(p2, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    try:
        re3 = re.search(p3, str(gangseo.loc[idx, '세부 주소'])).group()
        re3 = re3.group(1)+re3.group(2)+' '
    except:
        try:
             re3 = re.search(p4, str(gangseo.loc[idx, '세부 주소'])).group()+' '
        except:
            re3 = re.search(p5, str(gangseo.loc[idx, '세부 주소'])).group()+' '
        

    
    gangseo.loc[idx, '세부 주소_clean'] = re1+re2+re3
    idx += 1
gangseo
```

```python
#위도 경도를 추출해주는 사용자 정의 함수 만들기

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x = geo.latitude 
    y = geo.longitude
    return x, y
```

```python
#위도, 경도 컬럼 만들어주기

idx =0
for i in gangseo['세부 주소_clean']:
    try:
        gangseo.loc[idx,'위도'] = geocoding(i)[0]
        gangseo.loc[idx,'경도'] = geocoding(i)[1]
    except:
        pass
    idx += 1
gangseo
```
