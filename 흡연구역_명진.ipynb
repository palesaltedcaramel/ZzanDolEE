#### 칼럼명 다시 지정

```python
districts = [songpa, seodaemun, jungnang, dongdaemun, nowon] #마포, 강서 파일 열기 에러 나서 아직 데이터프레임 생성 못함 
for district in districts:
    print(district.columns)
```

```python
#필요없는 칼럼 삭제
songpa.drop(['연번', '데이터기준일자'], axis = 1, inplace = True)
seodaemun.drop(['시설구분', '설치위치 상세', '규모(㎡)', '설치일', '설치주체', '관리여부', '관리기관'], axis = 1, inplace = True)
jungnang.drop(['업종', '흡연실 개소수'], axis = 1, inplace = True)
dongdaemun.drop(['규모', '설치년도', '운영관리', '데이터기준일자'], axis = 1, inplace = True)
nowon.drop(['규모', '설치일', '설치주체', '운영관리', '기준일자'], axis = 1, inplace = True)
```

```python
for district in districts:
    print(district.columns)
```

```python
#없는 칼럼 추가
songpa['지역(행정구))']='송파구'
jungnang['지역(행정구))']='중랑구'
dongdaemun['지역(행정구))']='동대문구'
nowon['지역(행정구))']='노원구'
```

```python
#칼럼이름 재지정
songpa.rename(columns = {'구분':'흡연시설형태','도로명주소:''세부 주소'}, inplace=True)
seodaemun.rename(columns = {'자치구':'지역(행정구', '시설형태':'흡연시설형태','설치위치:''세부 주소'}, inplace=True)
jungnang.rename(columns = {'시설명(업소)':'건물/상호명'}, inplace=True)
dongdaemun.rename(columns= {'설치주체':'건물/상호명'}, inplace=True)
nowon.rename(columns = {'시설형태':'흡연시설형태'}, inplace=True)
```
