```python
#위도 경도를 추출해주는 사용자 정의 함수 만들기
!pip install geopy
import geopy
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x = geo.latitude 
    y = geo.longitude
    return x, y
```

```python
#위도, 경도 채우기
for district in districts:
    idx =0
    for i in district['세부 주소']:
        try:
            district.loc[idx,'위도'] = geocoding(i)[0]
            district.loc[idx,'경도'] = geocoding(i)[1]
        except:
            pass
        idx += 1
    district
```
