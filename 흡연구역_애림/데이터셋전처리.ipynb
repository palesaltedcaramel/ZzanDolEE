#### 강서구

```python
#지역구 데이터 불러오기

gangseo = pd.read_csv('./강서구/서울특별시 강서구_실외 흡연 장소_20201008.csv', encoding = 'euc-kr')
gangseo
```

```python
import re

#불필요한 컬럼 삭제
gangseo = gangseo.drop(['시설 구분', '규모', '설치일', '관리여부', '관리'], axis = 1)

#컬럼 내용 표준화: 흡연시설형태
gangseo['시설형태'] = gangseo['시설형태'].replace('완전' ,'', regex = True)
gangseo

#컬럼명 변경
gangseo.rename(columns = {'자치구' : '지역(행정구)','시설형태':'흡연시설형태', '설치 위치': '세부 주소',  '설치 주체': '건물/상호명'}, inplace = True)

#주소 다듬기: 정확한 주소 표기

idx = 0
for i in list(gangseo[ '세부 주소']):
    i = '서울특별시 강서구 ' + str(i)
    gangseo.loc[idx, '세부 주소'] = i
    idx += 1
    
gangseo
```

```python
#주소 다듬기: 주소 표기 정규표현식으로 수정

idx = 0
for i in list(gangseo[ '세부 주소']):
    p1 = re.compile('([\w]+[시])')
    p2 = re.compile('[\w]+[구]')
    p3 = re.compile ('([\w]+[로]) ([\w]+[길] [\d]+)')       # ~로~길이 붙어있음;;;;
    
    p4 = re.compile ('[\w]+[로] [\d]+')
    p5 = re.compile ('[\w]+[길] [\d]+')

    re1 = re.search(p1, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    re2 = re.search(p2, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    try:
        re3 = re.search(p3, str(gangseo.loc[idx, '세부 주소'])).group()
        re3 = re3.group(1)+re3.group(2)+' '
    except:
        try:
             re3 = re.search(p4, str(gangseo.loc[idx, '세부 주소'])).group()+' '
        except:
            re3 = re.search(p5, str(gangseo.loc[idx, '세부 주소'])).group()+' '
        

    
    gangseo.loc[idx, '세부 주소_clean'] = re1+re2+re3
    idx += 1
gangseo
```

```python
# !pip install geopy
```

```python
#위도 경도를 추출해주는 사용자 정의 함수 만들기

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x = geo.latitude 
    y = geo.longitude
    return x, y
```

```python
#컬럼 생성: 위도, 경도

idx =0
for i in gangseo['세부 주소_clean']:
    try:
        gangseo.loc[idx,'위도'] = geocoding(i)[0]
        gangseo.loc[idx,'경도'] = geocoding(i)[1]
    except:
        pass
    idx += 1
gangseo
```

```python
#전처리된 데이터프레임 새로 만들기

gangseo_clean = gangseo[['지역(행정구)', '건물/상호명','흡연시설형태', '세부 주소', '위도', '경도']]
gangseo_clean 
```

```python
#건물/상호명 다듬기

idx = 0
for i in sungdong['설치위치']:
    i = re.sub('\([\w ]+\)','',str(i))
    i = i.split(' ')[0]
    
    #정제된 위치명을가지고 새로운 변수 생성
    sungdong.loc[idx,'설치위치_clean'] = i
    idx += 1

sungdong
```

```python
#건물/상호명에 따른 주소 크롤링

import os
import sys
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome('C:/cakd7/tool/chromedriver.exe')

idx = 0
for i in sungdong['설치위치_clean']:
    driver.get('https://map.kakao.com/')
    time.sleep(1)

    
    #카카오맵에서 검색
    search = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')
    search.send_keys(i)
    search.send_keys(Keys.ENTER)
    time.sleep(1)
    
    #주소 크롤링
    address = driver.find_element(By.XPATH, '/html/body/div[5]/div[2]/div[1]/div[7]/div[4]/ul/li[1]/div[5]/div[2]/p[1]')
    sungdong.loc[idx, '세부 주소'] = address.text
    
    idx += 1
    
    
driver.close()
```

```python
#컬럼 생성: 위도, 경도

idx =0
for i in sungdong['세부 주소']:
    try:
        sungdong.loc[idx,'위도'] = geocoding(i)[0]
        sungdong.loc[idx,'경도'] = geocoding(i)[1]
    except:
        pass
    idx += 1
sungdong
```

```python
#컬럼명 변경

sungdong.rename(columns = {'구분': '흡연시설형태', '설치위치': '건물/상호명'},  inplace = True)
sungdong
```

#### 영등포구

```python
#지역구 데이터 불러오기

yeong = pd.read_csv('영등포구/서울특별시 영등포구_흡연구역_실외_ 현황_20220429.csv', encoding = 'euc-kr')
yeong
```

```python
#불필요한 컬럼 삭제

yeong.drop(columns = ['설치 위치', '규모(제곱미터)', '설치일', '설치 주체','관리여부','관리'], axis = 1, inplace = True)
yeong
```

```python
#주소 다듬기: 정확한 주소 표기


idx = 0
for i in yeong['시설 구분']:
    i = str(i).replace('㈜','')
    i = re.sub('㈜|앞|옆|뒤|A,B동|쪽','',i)
    yeong.loc[idx, '시설 구분_clean'] = i
    idx += 1
yeong
```

```python
#컬럼 내용 표준화: 흡연시설형태

yeong = yeong.replace('완전','', regex = True)
yeong
```

```python
#건물/상호명에 따른 주소 크롤링

driver = webdriver.Chrome('C:/cakd7/tool/chromedriver.exe')

idx = 0
for i in yeong['시설 구분_clean']:
    driver.get('https://map.kakao.com/')
    time.sleep(1)

    
    #카카오맵에서 검색
    search = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')
    search.send_keys('영등포 ' + i)
    search.send_keys(Keys.ENTER)
    time.sleep(1)
    
    #주소 크롤링
    try:
        address = driver.find_element(By.CSS_SELECTOR, '#info\.search\.place\.list > li.PlaceItem.clickArea.PlaceItem-ACTIVE > div.info_item > div.addr > p:nth-child(1)')
        yeong.loc[idx, '세부 주소'] = address.text
    except:
        try:
            address = driver.find_element(By.CSS_SELECTOR, '#info\.search\.place\.list > li:nth-child(1) > div.info_item > div.addr > p:nth-child(1)')
            yeong.loc[idx, '세부 주소'] = address.text
        except:
            address = driver.find_element(By.CSS_SELECTOR, '#info\.search\.place\.list > li:nth-child(2) > div.info_item > div.addr > p:nth-child(1)')
            yeong.loc[idx, '세부 주소'] = address.text
            
    idx += 1
    
    
driver.close()
```

```python
#컬럼명 변경

yeong = yeong.rename(columns = {'자치구': '지역(행정구)', '시설 구분':'건물/상호명','시설형태':'흡연시설형태'})
yeong
```
