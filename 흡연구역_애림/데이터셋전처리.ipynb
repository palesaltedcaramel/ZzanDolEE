```python
#지역구 데이터 불러오기

gangseo = pd.read_csv('./강서구/서울특별시 강서구_실외 흡연 장소_20201008.csv', encoding = 'euc-kr')
gangseo
```

```python
import re

#불필요한 컬럼 삭제
gangseo = gangseo.drop(['시설 구분', '규모', '설치일', '관리여부', '관리'], axis = 1)

#컬럼 내용 표준화: 흡연시설형태
gangseo['시설형태'] = gangseo['시설형태'].replace('완전' ,'', regex = True)
gangseo

#컬럼명 변경
gangseo.rename(columns = {'자치구' : '지역(행정구)','시설형태':'흡연시설형태', '설치 위치': '세부 주소',  '설치 주체': '건물/상호명'}, inplace = True)

#주소 다듬기: 정확한 주소 표기

idx = 0
for i in list(gangseo[ '세부 주소']):
    i = '서울특별시 강서구 ' + str(i)
    gangseo.loc[idx, '세부 주소'] = i
    idx += 1
    
gangseo
```

```python
#주소 다듬기: 주소 표기 정규표현식으로 수정

idx = 0
for i in list(gangseo[ '세부 주소']):
    p1 = re.compile('([\w]+[시])')
    p2 = re.compile('[\w]+[구]')
    p3 = re.compile ('([\w]+[로]) ([\w]+[길] [\d]+)')       # ~로~길이 붙어있음;;;;
    
    p4 = re.compile ('[\w]+[로] [\d]+')
    p5 = re.compile ('[\w]+[길] [\d]+')

    re1 = re.search(p1, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    re2 = re.search(p2, str(gangseo.loc[idx, '세부 주소'])).group()+' '
    try:
        re3 = re.search(p3, str(gangseo.loc[idx, '세부 주소'])).group()
        re3 = re3.group(1)+re3.group(2)+' '
    except:
        try:
             re3 = re.search(p4, str(gangseo.loc[idx, '세부 주소'])).group()+' '
        except:
            re3 = re.search(p5, str(gangseo.loc[idx, '세부 주소'])).group()+' '
        

    
    gangseo.loc[idx, '세부 주소_clean'] = re1+re2+re3
    idx += 1
gangseo
```

```python
#셀레니움으로 구글어스에서 위도, 경도 크롤링
import selenium
from selenium  import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

path = 'c:/cakd7/tool/chromedriver.exe'
url = 'https://earth.google.com/web/search/'

driver = webdriver.Chrome(path)


#위도값이 null값인 행의 인덱스를 추출하여, 세부주소로 위도, 경도 크롤링 하기
for i in aerim_df[aerim_df['위도'].isnull()].index:
    j = aerim_df.loc[i,'세부 주소_clean']
    driver.get(url+j)
    time.sleep(5)
    try:
        now_url = driver.current_url
        split_url = now_url.split('@')[1]
        split_url = split_url.split(',')
        aerim_df.loc[i, '위도'] = split_url[0]
        aerim_df.loc[i, '경도'] = split_url[1]
    except:
        pass

driver.close()
```

```python
#위도 경도를 추출해주는 사용자 정의 함수 만들기

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x = geo.latitude 
    y = geo.longitude
    return x, y
```

```python
#컬럼 생성: 위도, 경도

idx =0
  
for i in aerim_df[aerim_df['위도']== '0'].index:
    address = aerim_df.loc[i, '세부 주소_clean']
    try:
        aerim_df.loc[idx,'위도'] = geocoding(address)[0]
        aerim_df.loc[idx,'경도'] = geocoding(address)[1]
    except:
        pass
    idx += 1
aerim_df
```
