# 용산구 

````python
import pandas as pd

yongsan= pd.read_csv('흡연구역_주희/서울특별시 용산구_흡연구역_10_22_2021.csv', encoding='cp949')
yongsan
````

````python
yongsan.drop(['시설 구분'], axis = 1, inplace = True)
````

````python
yongsan.rename(columns = {'자치구명' : '지역(행정구)', '시설형태':'흡연시설형태', '설치 주체':'건물/상호명', '서울특별시 용산구 설치 위치':'세부 주소'}, inplace = True)
````

````python
yongsan.columns
````

````python
yongsan = yongsan[['지역(행정구)','건물/상호명', '흡연시설형태', '세부 주소', '위도', '경도']]
yongsan
````

````python
import re
idx = 0

for i in yongsan['건물/상호명']:
    i = re.sub('\W','',i)
    yongsan.loc[idx, '건물/상호명'] = i
    idx += 1
yongsan
````
