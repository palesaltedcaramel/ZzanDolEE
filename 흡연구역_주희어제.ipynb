# 서초구

````python
import pandas as pd

seocho= pd.read_csv('흡연구역_주희/서울특별시 서초구_흡연시설 현황_20220520.csv', encoding='cp949')
seocho
````

````python
#자치구명 수정 
seocho['자치구명'] = seocho['자치구명'].str.replace('서초구청', '서초구')
seocho.head()
````

````python
#시설형태 수정
seocho['시설형태'].unique()
````

````python
#개방형흡연부스, 라인형흡연구역, 부분개방형흡연실, 개방형흡연구역, 부분개방형흡연부스, 부분개방형흡연구역 > 개방형 

import re
idx = 0
for i in seocho['시설형태']:
    i =  re.sub('[가-힣]+[개방|라인]+[형]+[가-힣]+', '개방형', i)
    seocho.loc[idx, '시설형태'] = i
    idx += 1
seocho['시설형태']
````

````python
 #폐쇄형흡연부스, 비대면흡연부스 > 폐쇄형 
import re
idx = 0
for i in seocho['시설형태']:
    i =  re.sub('[비대면|폐쇄형]+[가-힣]+', '폐쇄형', i)
    seocho.loc[idx, '시설형태'] = i
    idx += 1
seocho['시설형태']
````

````python
seocho
````

````python
seocho.drop(['설치일자'], axis = 1, inplace = True)
````

````python
seocho.rename(columns = {'자치구명' : '지역(행정구)', '시설형태':'흡연시설형태', '설치주체':'건물/상호명', '설치위치':'세부 주소'}, inplace = True)
seocho = seocho[['지역(행정구)','건물/상호명', '흡연시설형태', '세부 주소']]
seocho.head()
````

````python
#세부 주소 전처리
import re 
import numpy as np
idx = 0
for i in list(seocho['세부 주소']):
    p1 = re.compile ('[\w]+[역]\s?\d[번]\s?[출구]+') 
    p2 = re.compile ('[\w]{,2}[동]\s?\d+[-]\d+')
    p3 = re.compile ('[\w]{,2}[동]\s?\d+')
    p4 = re.compile ('[\w]+[로][\w]+[길]\s?[\d]+')  
    p5 = re.compile ('[\w]+[로]\s?[\d]+')
    p6 = re.compile ('[센트럴시티]+\s\w{,2}[선]')
    p7 = re.compile ('[\w㈜]+')
    try:
        re1 = re.search(p1, str(seocho.loc[idx, '세부 주소'])).group()
        
    except:
        try: 
            re1 = re.search(p2, str(seocho.loc[idx, '세부 주소'])).group()
            
        except:
            try:
                re1 = re.search(p3, str(seocho.loc[idx, '세부 주소'])).group()
            except:
                try:
                    re1 = re.search(p4, str(seocho.loc[idx, '세부 주소'])).group()
                except:
                    try:
                        re1 = re.search(p5, str(seocho.loc[idx, '세부 주소'])).group()
                    except:
                        try:
                            re1 = re.search(p6, str(seocho.loc[idx, '세부 주소'])).group()
                        
                        except:
                            re1 = re.match(p7, str(seocho.loc[idx, '세부 주소'])).group()

    
    seocho.loc[idx, '세부 주소_clean'] = re1
    
    idx += 1
    

seocho
````

````python
seocho['위도'] = np.nan
seocho['경도'] = np.nan
seocho
````
