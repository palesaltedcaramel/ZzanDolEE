# 은평구

````python
import pandas as pd

eunpyeong= pd.read_csv('흡연구역_주희/서울특별시_은평구_흡연실현황_20220101.csv', encoding='cp949')
eunpyeong
````

````python
#소재지도로명주소 전처리
import re 
idx = 0
for i in list(eunpyeong['소재지도로명주소']):
    p1 = re.compile ('([\w]+[로])([\w]+[길] [\d]+)') 
    p2 = re.compile ('([\w]+[로])([\w]+[길] [\d]+[-][\d]+)')
    p3 = re.compile ('[\w]+[로] [\d]+[-][\d]+')
    p4 = re.compile ('[\w]+[길] [\d]+[-][\d]+')
    p5 = re.compile ('[\w]+[로] [\d]+')
    p6 = re.compile ('[\w]+[길] [\d]+')
    
    try:
        re1 = re.search(p1, str(eunpyeong.loc[idx, '소재지도로명주소'])).group()
        re1 = re1.group(1)+re1.group(2)
    except:
        try: 
            re1 = re.search(p2, str(eunpyeong.loc[idx, '소재지도로명주소'])).group()
            re1 = re1.group(1)+re1.group(2)
        except:
            try:
                re1 = re.search(p3, str(eunpyeong.loc[idx, '소재지도로명주소'])).group()
            except:
                try:
                    re1 = re.search(p4, str(eunpyeong.loc[idx, '소재지도로명주소'])).group()
                except:
                    try:
                        re1 = re.search(p5, str(eunpyeong.loc[idx, '소재지도로명주소'])).group()
                    except:
                        try:
                            re1 = re.search(p6, str(eunpyeong.loc[idx, '소재지도로명주소'])).group()
                        
                        except:
                            pass

    
    eunpyeong.loc[idx, '소재지도로명주소_clean'] = re1
    idx += 1
    

eunpyeong
````

````python
#위도 경도를 추출해주는 사용자 정의 함수 만들기

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x = geo.latitude 
    y = geo.longitude
    return x, y
````

````python

idx =0
for i in eunpyeong_a['세부 주소']:
    try:
        eunpyeong_a.loc[idx,'위도'] = geocoding(i)[0]
        eunpyeong_a.loc[idx,'경도'] = geocoding(i)[1]
    except:
        pass
    idx += 1
eunpyeong_a
````
